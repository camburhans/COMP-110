# Hypothesis
### There will be a very strong correlation between the number of deaths among people aged 15-19 and the probability of dying among people aged 15-19 years old.

### Importing pandas as pd

```python
import pandas as pd
```

### Initializing the two constants

```python
databank = pd.read_csv("../../data/databank_education_2018.csv")

databank = databank.rename(columns={"SH.DYN.1519": "death_prob_1519"})
databank = databank.rename(columns={"SH.DTH.1519": "num_deaths_1519"})
```

### Shaping and printing the DataBank, then showing the first ten columns

```python
print(databank.shape)
print(databank.head(10))
```

### Three columns that I am focusing in on.

```python
three_col = databank[["Country", "num_deaths_1519", "death_prob_1519"]]
print(three_col.head(10))
```

### Narrowing down the columns to only include the rows that have values

```python
col_notna1 = databank[databank["Country"].notna()]
col_notna2 = col_notna1[col_notna1["death_prob_1519"].notna()]
col_notna3 = col_notna2[col_notna2["death_prob_1519"].notna()]
col_notna = col_notna2[["Country", "death_prob_1519", "num_deaths_1519"]]
print(col_notna.shape)
print(col_notna.head(10))
```

### Printing the columns that show the ten highest-value rows of the death probabilities

```python
print(col_notna.sort_values(by=["death_prob_1519"], ascending=False).head(10))
```

### Scatter plot for the two constants I am looking at.

```python
col_notna.plot.scatter(x = "death_prob_1519", y = "num_deaths_1519", c = "Blue")
```

### Correlation number. The closer to 1 or -1, the more strongly correlated

```python
prob = col_notna["death_prob_1519"]
deaths = col_notna["num_deaths_1519"]
print(prob.corr(deaths))
```

# Conclusion
### There is not a high correlation at all between these two statistics. As much as I thought there would be a strong correlation, it turns out that there is no data to prove or support the statement that "There will be a very strong correlation between the number of deaths among people aged 15-19 and the probability of dying among people aged 15-19 years old.""
